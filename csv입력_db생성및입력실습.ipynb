# Write csv

```python
import csv

rows=[]

with open('data-text.csv',encoding = 'utf-8') as f:
    reader=csv.reader(f) #csv reader 객체 생성
    for row in reader:
        rows.append(row)
rows[:100]
```

```python
#csv write
with open('data-text20.csv','w',encoding='utf-8',newline='') as f:
    writer=csv.writer(f)
    writer.writerows(rows[:20]) #데이터 20개로 잘라냄
```
