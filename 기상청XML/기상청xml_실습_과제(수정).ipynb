```python
import xml.etree.ElementTree as ET

tree=ET.parse('mid-term-rss3.xml')
```

```python
root=tree.getroot()
```

```python
list_city= root.findall('channel/item/description/body/location/city')
list_date= root.findall('channel/item/description/body/location/data/tmEf')
list_wf=root.findall('channel/item/description/body/location/data/wf')
list_low=root.findall('channel/item/description/body/location/data/tmn')
list_max=root.findall('channel/item/description/body/location/data/tmx')
```

```python
cities=[]
for i in range(35):
    for j in range(13):
        cities.append(list_city[i].text)
        
        
```

```python
dates=[]
wfs=[]
lows=[]
maxes=[]
for date,wf,low,maxi in zip(list_date,list_wf,list_low,list_max):
    dates.append(date.text)
    wfs.append(wf.text)
    lows.append(low.text)
    maxes.append(maxi.text)
```

```python
i=1
for city in list_city:
    print(city.text,i)
    i+=1
```

```python
for j in range(526):
    print(cities[j],dates[j],wfs[j],lows[j],maxes[j])
```

```python
for i in range(525):
    if cities[i] != cities[i-1]:
        print('<',cities[i],'>')
    print(dates[i],wfs[i],'  ',lows[i],'~',maxes[i],'도')
```
