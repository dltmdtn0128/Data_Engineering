```python
import xml.etree.ElementTree as ET
tree=ET.parse('mid-term-rss3.xml')
```

```python
root= tree.getroot()
```

```python
for location in root.findall('channel/item/description/body/location'):
    city= location.find('city').text
    print('<',city,'>')
    for data in location.findall('data'):
        tmEf = data.find('tmEf').text
        wf= data.find('wf').text
        tmn = data.find('tmn').text
        tmx = data.find('tmx').text
        if len(wf)>2:
            print(tmEf,wf,'\t',tmn,' ~ ',tmx,'도')
        else:
            print(tmEf,wf,'\t\t',tmn,' ~ ',tmx,'도')
```
