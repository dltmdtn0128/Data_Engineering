```python
import seaborn as sns
```

```python
#load titanic dataset
df= sns.load_dataset('titanic')
df
```

```python
#summary 누락데이터 확인
df.info()
```

```python
#누락데이터 개수 구하기 1
#count NaN in deck column
nan_deck=df['deck'].value_counts(dropna=False) #dropna=False -> 누락데이터 개수 포함
nan_deck
```

```python
#누락데이터 개수 구하기 2
#find missing data
df.head().isnull()
```

```python
#count missing data using isnull()
df.head().isnull().sum(axis=0) #axis=0은 행방향/ axis=1은 열방향
```

```python
#누락데이터 개수 파악
missing_df=df.isnull()
for col in missing_df.columns:
    missing_count=missing_df[col].value_counts()
    try:
        print(col,' : ',missing_count[True])
    except:
        print(col,' : ',0)
```

```python
#난 이렇게 할래
df.isnull().sum(axis=0)
```

```python
#누락데이터가 많은 열 제거
df_thresh = df.dropna(axis=1,thresh=500) #axis=1 옵션 : 열제거 thresh =500 옵션 : nan값 개수 500개 이상 열 제거
df_thresh.columns
#500개 이상인 deck열만 제거됨
```

```python
#age열에 나이 데이터가 없는 모든 행 삭제
df_age = df.dropna(axis=0,subset=['age'],how='any') #age열에 대해서 how='any'옵션 : nan이 하나라도 존재하면 how='all'옵션은 모두 
len(df_age)
```

```python
#age 열 누락데이터 확인
df['age'].head(20) #20개행 출력 i[0~19]까지
```

```python
#평균값 계산
mean_age = df['age'].mean()
mean_age
```

```python
#평균값으로 nan값 대체해서 입력
df['age'].fillna(mean_age,inplace=True)#fillna 빈값 대체하기
df['age'].head(20)
```
