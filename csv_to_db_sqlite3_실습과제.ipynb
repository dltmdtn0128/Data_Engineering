```python
import csv
rows=[]

with open('data-text.csv',encoding = 'utf-8') as f:
    reader=csv.reader(f)
    for row in reader:
        rows.append(row)
rows[:20]
```

```python
with open('data-text20.csv','w',encoding='utf-8',newline='') as f: #csv파일을 20개로 자르기
    writer=csv.writer(f)
    writer.writerows(rows[:20])
```

```python
short=[]
with open('data-text20.csv','r',encoding='utf-8',newline='') as t: #20개로 줄인 csv파일을 db에 넣기 편하게 리스트에 집어넣기
    reader=csv.reader(t)
    for rows in reader:
        short.append(rows)
```

```python
import sqlite3
con=sqlite3.connect("C:/Users/Seung Soo/Desktop/Data Eng/data-text_db.db")#sqlite3 불러와 새로운 db파일 만들기
cursor=con.cursor()#커서객체생성
```

```python
cursor.execute("CREATE TABLE life_expectency(indicator text, country text, display_value int)")#db테이블 생성
```

```python
for row in short:
    cursor.execute('INSERT INTO life_expectency VALUES(?,?,?)',(row[0],row[5],row[7]))
    #indicator, county display value만 불러와야하니 0,5,7순서로 불러옴
```

```python
con.commit() #db에 반영
```

```python
con.close()#db연결 종료
```
