```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Seung Soo\\Desktop\\Data_engineering\\데이터전처리_2_수업\\auto-mpg.csv',header=None,delimiter=r"\s+")
```

```python
#리터당 키로미터 (km/l)로 변환하기
df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
df.head()
```

```python
#연비단위 환산
mpg2kpl = 1.60934/3.78541
mpg2kpl
```

```python
#kpl column 추가
df['kpl'] = df['mpg']*mpg2kpl
df['kpl'] = df['kpl'].round(2)
df.head()
```

```python
#데이터타입 확인
df.info()
```

```python
#왜 horsepower가 문자열?
print(df['horsepower'].unique())
#물음표로 데이터가 오염되어있다.
```

```python
#값 치환, ?을 제거하기위해 ?을 NaN값으로 변환
import numpy as np
df['horsepower'].replace('?',np.nan,inplace=True)
```

```python
df.info()
```

```python
df['horsepower'].unique()
```

```python
#누락데이터 제거
df.dropna(subset=['horsepower'],inplace = True)
df.info()
#392개로 돌아옴
```

```python
#astype() 로 자료형 변환 -> 문자열을 실수형으로 변환
df['horsepower'] = df['horsepower'].astype('float')
print(df['horsepower'].dtypes)
```

```python
#히스토그램
count, bin_dividers=np.histogram(df['horsepower'],bins=3)
print(count,bin_dividers)
#count는 각구간의 도수/빈도 // np.histogram은 최소값 46 / 최대값 230
#bin = 3개로 나누면 4개의 구간이 생성
```

```python
#cut()을 이용해서 구간분할하기
bin_names = ['저출력','보통출력','고출력']
df['hp_bin'] = pd.cut(df['horsepower'], bins=bin_dividers,labels = bin_names, include_lowest = True)
df[['horsepower','hp_bin']].head(15)
```

```python
#더미변수 구하기 (더미변수는 여부에 따라 0 또는 1로 표현)
horsepower_dummies= pd.get_dummies(df['hp_bin'])
horsepower_dummies.head(15)
```
