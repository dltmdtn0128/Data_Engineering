```python
#모든 누락데이터를 해당 해 전세계 GDP성장률 평균으로 대체
df=df.apply(lambda x: x.fillna(x.mean()),axis=0)
df
```

```python
#누락데이터 제거 완료
missing_df=df.isnull()
for col in missing_df.columns:
    missing_count=missing_df[col].value_counts()
    try:
        print(col,": ",missing_count[True])
    except:
        print(col,': ',0)
```

```python
year=1980
for min,max,std in zip(df.min(),df.max(),df.std()):
    print(year,'min : ',min,'\t',year,'max : ',max,'\t',year,'std : ',std)
    year+=1
```

```python
delta=2
bottom= df.mean()-delta*df.std()

top = df.mean()+delta*df.std()
print(bottom,top)
```

```python
bottom = []
top = []

for i in df.mean()-delta*df.std():
    bottom.append(i)
for i in df.mean()+delta*df.std():
    top.append(i)
```

```python
#2020 이후의 데이터는 예측치이기에 표준편차안에들어가는 값에 해당되어 값이 출력되지 않은걸 확인할 수있다. 
lists=[]

for i,b,t in zip(df.index,bottom,top):
    for c, val in zip(df.columns, df.loc[i]):
        if val<b:
            lists.append([c,i,val])
        elif val > t:
            lists.append([c,i,val])
```

```python
# 연도 및 국가별 이상치 확인
lists = sorted(lists)
for i in range(len(lists)):
    print('{}  {:30}  {}'.format(lists[i][0],lists[i][1],lists[i][2]))
```
