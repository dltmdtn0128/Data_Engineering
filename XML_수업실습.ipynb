```python
#해당모듈 사용
import xml.etree.ElementTree as ET

#parse xml file
tree=ET.parse('data-text.xml')
```

```python
#get root node
root=tree.getroot()

#root tag
#root.tag

#데이터 유형의 메소드 속성 확인 데이터 타입의 method,attrib list
dir(root)

#자식목록 확인 xml을 다 들어가서 보면 좋지만 보기 힘들기에 찾아서 보여줌
list(root)
```

```python
#Element 의 속성과 메소드
root.find('Data').tag
```

```python
#Tag의 text 추출
root.find('Disclaimer/Display').text #disclaimer밑에 display가 들어있는걸 찾는거
```

```python
#Tag의 attrib.추출
root.find('Data/Observation/Dim').attrib
```

```python
#findall()
list_dim = root.findall('Data/Observation/Dim')
print(list_dim[8].attrib)
```

```python
#모든 dim의 attrib 출력
for dim in list_dim:
    print(dim.attrib)
```

```python
#observation 을 하나의 list에 record 저장
observation=[]
list_obs=root.find('Data/Observation').attrib
for obs in list_obs:
    observation.append(list_obs)

observation
```
